```python
# Pfad zum Ordner mit den CSV-Dateien
folder_path = 'out/ready_csvs'
column_to_clean = "ID"


def format_value(value):
    try:
        if float(value).is_integer():
            return int(float(value))
        else:
            return value
    except ValueError:
        return value
    
    
# Durchlaufen aller Dateien im Ordner
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        
        # CSV-Datei einlesen
        data = pd.read_csv(file_path)

        print(data.columns)

        if "Value" in data.columns:
            data['Value'] = data['Value'].apply(format_value)

        if column_to_clean in data.columns:
            # Entfernen von Leerzeichen am Anfang und Ende der Strings in der Spalte
            data[column_to_clean] = data[column_to_clean].astype(str).str.strip()

        if 'Year' in data.columns:
            first_value = data['Year'].iloc[0]
            if isinstance(first_value, str) :
                data['Year'] = data['Year'].str.extract(r'(\d{4})')

        
        # CSV-Datei ohne Index speichern
        data.to_csv(file_path, index=False)

print("Alle CSV-Dateien wurden erfolgreich bearbeitet und gespeichert.")
```

```python
import pandas as pd

# Beispieldaten laden
data = pd.read_csv('out/ready_csvs/71487ned.csv')

# Kopie der Spalte 'Value' erstellen
value_column = data['Value'].copy()

# Konvertieren der Kopie in numerische Typen und dann in Integer-Werte
numeric_values = pd.to_numeric(value_column, errors='coerce')
value_column = numeric_values[~numeric_values.isna()].astype('Int64')

# Die ursprüngliche Spalte 'Value' durch die kopierte und formatierte Spalte ersetzen
data['Value'] = value_column

# Ergebnis anzeigen
data
data.to_csv('out/ready_csvs/71487ned.csv', index=False)


```
