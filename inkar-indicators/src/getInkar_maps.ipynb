```python
import pandas as pd
import geopandas as gpd
import mapclassify as mc
```

```python
metadata = pd.read_csv("out/csvs/indicator_Arbeitslosenquote.csv")
metadata
```

```python
#vg250_sta Flächen Ebene Staat
#vg250_lan Flächen Ebene Land
#vg250_krs Flächen Ebene Kreis
#vg250_gem
municipal_boundaries = gpd.read_file(filename='DE_VG250.gpkg', layer="vg250_krs",engine="pyogrio")
```

```python
municipal_boundaries.head(5)
```

```python
municipal_boundaries
```

```python
municipal_boundaries['AGS_int'] = municipal_boundaries.AGS_0.astype(int)
```

```python
metadata
```

```python
municipal_boundaries = pd.merge(municipal_boundaries, metadata,
                               left_on = "AGS_int", 
                               right_on = "Kennziffer")
```

```python
# Natural_Breaks, Equal_Interval, Quantiles, Standard_Deviation
#
# boxplot, equalinterval, fisherjenks, fisherjenkssampled, headtailbreaks, jenkscaspall, 
# jenkscaspallforced, jenkscaspallsampled, maxp, maximumbreaks, naturalbreaks, quantiles, 
# percentiles, prettybreaks, stdmean, userdefined

p = municipal_boundaries[municipal_boundaries['Zeitbezug']== 2000].plot(column='Wert',
                             scheme="fisherjenks",
                             k=5,
                             legend = True, 
                             figsize = (10,8))
p.axis('off')
p.set_title('Quantiles, 2020')
```
