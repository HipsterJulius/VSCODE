```python

data = pd.read_csv("out/new.csv", index_col="Unnamed: 0")
data.head()
```

```python
inkar_übersicht_raum = pd.read_excel("out\übersicht\Übersicht der Indikatoren.xlsx", sheet_name=0)
inkar_übersicht_ZOM = pd.read_excel("out\übersicht\Übersicht der Indikatoren.xlsx", sheet_name=1)
inkar_übersicht_SDG = pd.read_excel("out\übersicht\Übersicht der Indikatoren.xlsx", sheet_name=2)


inkar_übersicht_raum = inkar_übersicht_raum.replace(';', '!.', regex=True)
inkar_übersicht_raum = inkar_übersicht_raum.replace('\n', '', regex=True)

inkar_übersicht_ZOM = inkar_übersicht_ZOM.replace(';', '!.', regex=True)
inkar_übersicht_ZOM = inkar_übersicht_ZOM.replace('\n', '', regex=True)

inkar_übersicht_SDG = inkar_übersicht_SDG.replace(';', '!.', regex=True)
inkar_übersicht_SDG = inkar_übersicht_SDG.replace('\n', '', regex=True)

inkar_übersicht_raum.to_csv("inkar_übersicht_raum.csv", index=False)
inkar_übersicht_raum = pd.read_csv("inkar_übersicht_raum.csv")

inkar_übersicht_ZOM.to_csv("inkar_übersicht_ZOM.csv", index=False)
inkar_übersicht_ZOM = pd.read_csv("inkar_übersicht_ZOM.csv")

inkar_übersicht_SDG.to_csv("inkar_übersicht_SDG.csv", index=False)
inkar_übersicht_SDG = pd.read_csv("inkar_übersicht_SDG.csv")



```

```python
colums = ['INKAR 2022 – Indikatorenübersicht:   Raumbeobachtung Deutschland', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6']
inkar_übersicht_raum = inkar_übersicht_raum.loc[:, colums]


colums = ['INKAR 2022 – Indikatorenübersicht:   Sustainable Development Goals (SDG)', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6']
inkar_übersicht_SDG = inkar_übersicht_SDG.loc[:, colums]

colums = ['INKAR 2022 – Indikatorenübersicht:   Zentrale-Orte-Monitoring (ZOM)', 'Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6']
inkar_übersicht_ZOM = inkar_übersicht_ZOM.loc[:, colums]

inkar_übersicht_raum.head()
#inkar_übersicht_SDG.head()
#inkar_übersicht_ZOM.head()
```

```python
for index in inkar_übersicht_raum.index:
    if (index == 0 or index == 1 or index == 2):
        continue
    algorithmus = inkar_übersicht_raum['Unnamed: 4'][index]
    description = inkar_übersicht_raum['Unnamed: 5'][index]
    statistical_basis = inkar_übersicht_raum['Unnamed: 6'][index]
    name_description = inkar_übersicht_raum['INKAR 2022 – Indikatorenübersicht:   Raumbeobachtung Deutschland'][index]
    for i in data.index:
        name_de = data['name_de'][i]
        if(name_description == name_de):
            data['description'][i] = description
            data['statistical_basis'][i] = statistical_basis
            data['algorithmus'][i] = algorithmus

for index in inkar_übersicht_SDG.index:
    if (index == 0 or index == 1 or index == 2):
        continue
    algorithmus = inkar_übersicht_SDG['Unnamed: 4'][index]
    description = inkar_übersicht_SDG['Unnamed: 5'][index]
    statistical_basis = inkar_übersicht_SDG['Unnamed: 6'][index]
    name_description = inkar_übersicht_SDG['INKAR 2022 – Indikatorenübersicht:   Sustainable Development Goals (SDG)'][index]
    for i in data.index:
        name_de = data['name_de'][i]
        if(name_description == name_de):
            data['description'][i] = description
            data['statistical_basis'][i] = statistical_basis
            data['algorithmus'][i] = algorithmus

for index in inkar_übersicht_ZOM.index:
    if (index == 0 or index == 1 or index == 2):
        continue
    algorithmus = inkar_übersicht_ZOM['Unnamed: 4'][index]
    description = inkar_übersicht_ZOM['Unnamed: 5'][index]
    statistical_basis = inkar_übersicht_ZOM['Unnamed: 6'][index]
    name_description = inkar_übersicht_ZOM['INKAR 2022 – Indikatorenübersicht:   Zentrale-Orte-Monitoring (ZOM)'][index]
    for i in data.index:
        name_de = data['name_de'][i]
        if(name_description == name_de):
            data['description'][i] = description
            data['statistical_basis'][i] = statistical_basis
            data['algorithmus'][i] = algorithmus

data.head()
```
