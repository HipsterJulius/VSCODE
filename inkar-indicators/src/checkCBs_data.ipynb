```python
import pandas as pd
import numpy as np
import cbsodata

data_netherlands = pd.DataFrame(cbsodata.get_meta('37201', 'TypedDataSet'))
#37201, 03747
data_netherlands
```

```python
data_nl = data_netherlands[['ID', 'RegioS','Perioden','LevendGeborenKinderenRelatief_2']]
#data_nl = data_netherlands[['ID', 'RegioS','Perioden','MannenEnVrouwen_4']]

#data_nl = data_nl[data_nl['Perioden'] == '2020JJ00']
data_nl['Perioden'] = data_nl['Perioden'].str.extract(r'(\d{4})')

#data_nl.replace(0, np.nan, inplace=True)
#data_nl = data_nl.dropna()

data_nl
```

```python
# Pivot-Tabelle erstellen
data_nl_years = data_nl.pivot(index='RegioS', columns='Perioden', values='LevendGeborenKinderenRelatief_2')

data_nl_years.reset_index(inplace=True)
data_nl_years.columns.name = None

#data_nl_years.to_csv("C:/Users/juliu/VSCODE/VSCODE/inkar-indicators/src/out/data_nl_years_nanfree.csv")
data_nl_years.to_excel("C:/Users/juliu/VSCODE/VSCODE/inkar-indicators/src/out/data_nl_years_.xlsx", index=False)
data_nl_years
```
