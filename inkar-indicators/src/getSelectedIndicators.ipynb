```python

import pandas as pd
import numpy as np
```

```python

data = pd.read_csv("out/translated_data.csv", index_col="Unnamed: 0")
data.head()
```

```python
indicators = pd.DataFrame(columns=['category', 'name', 'source', 'source_id'])  # Erstellen Sie ein leeres DataFrame mit der gewünschten Spalte

for index in data.index:
    if data['select'][index] == "x":
        print(index)
        category_value = data['category_de'][index]
        indicator_name = data['name_de'][index]
        source_value = data['source_de'][index]
        source_id_value = int(data['source_id_de'][index])
        indicators.loc[len(indicators)] = {'category': category_value, 'name': indicator_name, 'source': source_value, 'source_id':source_id_value}


# Optional: Wenn Sie den Index neu sortieren möchten
indicators.reset_index(drop=True, inplace=True)
indicators
```

```python
inkar = pd.read_csv("../data/inkar_2022.csv",sep=";",decimal=",",dtype={'Kennziffer':str,'Kennziffer_EU':str,'Name':str}, low_memory=False)
inkar.head()
```

```python
filtered_inkar = inkar[(inkar['Bereich'].isin(indicators['category'])) & (inkar['ID'].isin(indicators['source_id'])) & (inkar['Indikator'].isin(indicators['name']))]
filtered_inkar.reset_index(drop=True, inplace=True)
filtered_inkar

#filtered_inkar.to_csv("C:/Users/juliu/VSCODE/VSCODE/inkar-indicators/src/out/test2.csv")
```
