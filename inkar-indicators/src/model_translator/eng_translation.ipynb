```python
from transformers import MarianMTModel, MarianTokenizer
import pandas as pd
# Laden des Modells und des Tokenizers
model_name = "Helsinki-NLP/opus-mt-de-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)

# Text, den Sie übersetzen möchten
german_text = "Zuzüge"

# Tokenisieren und übersetzen
inputs = tokenizer(german_text, return_tensors="pt", padding=True, truncation=True)
outputs = model.generate(**inputs)

# Die generierten Ausgaben decodieren
translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(translated_text)
```

```python
data = pd.read_csv("..\out\data.csv",low_memory=False)
selected_columns = ['category_de', 'name_de', 'algorithmus', 'description', 'statistical_basis']
data = pd.concat([data[col] for col in selected_columns], axis=1)

data = data.assign(
    category_de=data['category_de'],
    name_de=data['name_de'],
    algorithmus_de=data['algorithmus'],
    description_de=data['description'],
    statistical_basis_de=data['statistical_basis'],
    category_en = "",
    name_en = "",
    algorithmus_en="",
    description_en="",
    statistical_basis_en="",
    )

column_order = ['category_de','name_de', "algorithmus_de", "description_de", "statistical_basis_de", 'category_en', 'name_en', "algorithmus_en", "description_en", "statistical_basis_en"]
data = data[column_order]
data.head()
```

```python
def ersetze_umlaute(text):
    text = text.replace('ü', 'ue')
    text = text.replace('ä', 'ae')    
    text = text.replace('ö', 'oe')
    return text
```

```python
for index in data.index:
    category_de = data['category_de'][index]
    print("Teil 1 : " + str(index))
    input_ids = tokenizer.encode(category_de, return_tensors="pt")
    output = model.generate(input_ids)
    translation = tokenizer.decode(output[0], skip_special_tokens=True)
    data['category_en'][index] = translation

print("Fertig 1")

for index in data.index:
    category_de = data['name_de'][index]
    print("Teil 2 : " + str(index))
    input_ids = tokenizer.encode(category_de, return_tensors="pt")
    output = model.generate(input_ids)
    translation = tokenizer.decode(output[0], skip_special_tokens=True)
    data['name_en'][index] = translation

print("Fertig 2")


for index in data.index:
    category_de = data['algorithmus_de'][index]
    print("Teil 3 : " + str(index))
    if(type(category_de) == str):
        #category_de = ersetze_umlaute(category_de)
        input_ids = tokenizer.encode(category_de, return_tensors="pt")
        output = model.generate(input_ids)
        translation = tokenizer.decode(output[0], skip_special_tokens=True)
        data['algorithmus_en'][index] = translation
    else: 
        data['algorithmus_en'][index] = ""

print("Fertig 3")



for index in data.index:
    category_de = data['statistical_basis_de'][index]
    print("Teil 4 : " + str(index))
    if(type(category_de) == str):
        input_ids = tokenizer.encode(category_de, return_tensors="pt")
        output = model.generate(input_ids)
        translation = tokenizer.decode(output[0], skip_special_tokens=True)
        data['statistical_basis_en'][index] = translation
    else: 
        data['statistical_basis_en'][index] = ""

print("Fertig 4")


for index in data.index:
    category_de = data['description_de'][index]
    print("Teil 5 : " + str(index))
    if(type(category_de) == str):
        input_ids = tokenizer.encode(category_de, return_tensors="pt")
        output = model.generate(input_ids)
        translation = tokenizer.decode(output[0], skip_special_tokens=True)
        data['description_en'][index] = translation
    else: 
        data['description_en'][index] = ""

print("Fertig 5")
```
