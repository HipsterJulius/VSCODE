```python
import pandas as pd
import cbsodata # type: ignore
data = pd.DataFrame(
        cbsodata.get_data('85826NED', 
                          filters="Perioden eq '2022JJ00'",
                          select=['RegioS', 'Werkloosheidspercentage_13', 'Perioden', 'Persoonskenmerken'])
                          )
data = data[data['Persoonskenmerken'] == "Totaal personen"]
data
```

```python
metadata = pd.DataFrame(cbsodata.get_meta('85826NED', 'TypedDataSet'))
```

```python
metadata = metadata[['RegioS', 'Werkloosheidspercentage_13', 'Perioden', 'Persoonskenmerken']]
metadata = metadata[metadata['Perioden'] == "2022JJ00"]
metadata = metadata[metadata['Persoonskenmerken'] == "T009002"]
metadata
```

```python
nl_regios_mapping = data
nl_regios_mapping.index = metadata.index
nl_regios_mapping = pd.concat([nl_regios_mapping, metadata], axis=1)


nl_regios_mapping = nl_regios_mapping[['RegioS']]
nl_regios_mapping.to_csv("NL_regionID_mapping.csv", index=False)
```
