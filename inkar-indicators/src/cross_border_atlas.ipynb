```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cross_border_atlas = pd.read_csv("cross_border_atlas_new.csv", sep=";")
cross_border_atlas
```

```python
# Balkendiagramm für Population
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 12))

# Population Balkendiagramm
axes[0].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Population min'], color='lightblue', label='Population min')
axes[0].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Population max'], left=cross_border_atlas['Population min'], color='blue', label='Population max')
axes[0].set_xlabel('Population')
axes[0].set_title('Population Range by Geographic Feature')
axes[0].legend()

# Area Balkendiagramm
axes[1].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Area min'], color='lightgreen', label='Area min')
axes[1].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Area max'], left=cross_border_atlas['Area min'], color='green', label='Area max')
axes[1].set_xlabel('Area (km²)')
axes[1].set_title('Area Range by Geographic Feature')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Balkendiagramm für Population
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 12))

# Population Balkendiagramm
axes[0].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Population min'], color='lightblue', height=0.4, align='center', label='Population min')
axes[0].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Population max'], color='blue', height=0.4, align='edge', label='Population max')
axes[0].set_xlabel('Population')
axes[0].set_title('Population Range by Geographic Feature')
axes[0].legend()

# Area Balkendiagramm
axes[1].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Area min'], color='lightgreen', height=0.4, align='center', label='Area min')
axes[1].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Area max'], color='green', height=0.4, align='edge', label='Area max')
axes[1].set_xlabel('Area (km²)')
axes[1].set_title('Area Range by Geographic Feature')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
# Balkendiagramm für Population und Area
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(14, 12))

# Population Balkendiagramm
axes[0].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Population min'], color='lightblue', height=0.4, align='center', label='Population min')
axes[0].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Population max'], color='blue', height=0.4, align='edge', label='Population max')
axes[0].set_xlabel('Population')
axes[0].set_xscale('log')
axes[0].set_title('Population Range by Geographic Feature')
axes[0].legend()

# Area Balkendiagramm
axes[1].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Area min'], color='lightgreen', height=0.4, align='center', label='Area min')
axes[1].barh(cross_border_atlas['geographic feature'], cross_border_atlas['Area max'], color='green', height=0.4, align='edge', label='Area max')
axes[1].set_xlabel('Area (km²)')
axes[1].set_xscale('log')
axes[1].set_title('Area Range by Geographic Feature')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
cross_border_atlas_population_min = cross_border_atlas.sort_values(by='Population min')
cross_border_atlas_population_max = cross_border_atlas.sort_values(by='Population max')
cross_border_atlas_area_min = cross_border_atlas.sort_values(by='Area min')
cross_border_atlas_area_max = cross_border_atlas.sort_values(by='Area max')

# Balkendiagramm für Population und Area
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 14))

# Population min Balkendiagramm
axes[0, 0].barh(cross_border_atlas['geographic feature'], cross_border_atlas_population_min['Population min'], color='lightblue')
axes[0, 0].set_xlabel('Population')
#axes[0, 0].set_xscale('log')
axes[0, 0].set_title('Population Min by Geographic Feature')

# Population max Balkendiagramm
axes[0, 1].barh(cross_border_atlas['geographic feature'], cross_border_atlas_population_max['Population max'], color='blue')
axes[0, 1].set_xlabel('Population')
#axes[0, 1].set_xscale('log')
axes[0, 1].set_title('Population Max by Geographic Feature')

# Area min Balkendiagramm
axes[1, 0].barh(cross_border_atlas['geographic feature'], cross_border_atlas_area_min['Area min'], color='lightgreen')
axes[1, 0].set_xlabel('Area (km²)')
#axes[1, 0].set_xscale('log')
axes[1, 0].set_title('Area Min by Geographic Feature')

# Area max Balkendiagramm
axes[1, 1].barh(cross_border_atlas['geographic feature'], cross_border_atlas_area_max['Area max'], color='green')
axes[1, 1].set_xlabel('Area (km²)')
#axes[1, 1].set_xscale('log')
axes[1, 1].set_title('Area Max by Geographic Feature')

plt.tight_layout()
plt.show()
```
